Importing relevant functionalities

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame as df
```

Preparing plots

```python
SVPD = pd.read_csv('../csv_files/Nele_Nstates_sumfk_SVPD.csv',header=0)
SVP = pd.read_csv('../csv_files/Nele_Nstates_sumfk_SVP.csv',header=0)
M321G = pd.read_csv('../csv_files/Nele_Nstates_sumfk_321G.csv',header=0)
ZINDO = pd.read_csv('../csv_files/Nele_Nstates_sumfk_ZINDO.csv',header=0)
TZVP = pd.read_csv('../csv_files/Nele_Nstates_sumfk_TZVP.csv',header=0)

plt.grid(True, linewidth=0.5, linestyle='--', axis = 'y',zorder=0)
# Make the plot
plt.scatter(ZINDO.N_states[:],ZINDO.sum_f_k[:],label='ZINDO',s=40, facecolors='none', edgecolors='seagreen')
plt.scatter(M321G.N_states[:],M321G.sum_f_k[:],label='3-21G',s=40, facecolors='none', edgecolors='yellow')
plt.scatter(SVP.N_states[:],SVP.sum_f_k[:],label='def2-SVP',s=40, facecolors='none', edgecolors='cyan')
plt.scatter(SVPD.N_states[:],SVPD.sum_f_k[:],label='def2-SVPD',s=40, facecolors='none', edgecolors='blue')
plt.scatter(TZVP.N_states[:],TZVP.sum_f_k[:],label='def2-TZVP',s=40, facecolors='none', edgecolors='red')

# Adding Xticks
plt.xlabel('States', fontsize = 15)
plt.ylabel('$\sum_k f_k$',fontsize = 15) #fontweight ='bold'
plt.ylim(0,80)
plt.xlim(0,7000)
plt.legend()
plt.show()
```
